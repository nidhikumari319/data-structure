# 2. Explain the key features of lists in Python.

ans:

1.Mutable: You can modify the list after its creation.

2.Ordered: Lists maintain the order of elements.

3.Heterogeneous: A list can contain elements of different types (e.g., integers, strings, objects).

4.Dynamic sizing: The size of the list can be changed as elements are added or removed.

5.Indexing and slicing: Lists support indexing and slicing operations.
